# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

def ReLU(x):
    c1 = x > 0
    c2 = abs(x)
    return c1*c2

def dReLU(x):
    return x>0

x = plt.linspace(-3,3,60)
plt.plot(x,ReLU(x))
plt.plot(x,dReLU(x))
plt.grid()
plt.show()


'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
```
